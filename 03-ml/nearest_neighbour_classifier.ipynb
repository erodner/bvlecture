# Nearest Neigbour classification

In this example, we are using the CIFAR-10 dataset with small 32x32 images.
This dataset is obtained from OpenML, which also gives us the following description of the
dataset:
```
Each instance represents a 32x32 colour image as a 3072-value array. The first 1024 entries contain the red channel values, the next 1024 the green, and the final 1024 the blue. The image is stored in row-major order, so that the first 32 entries of the array are the red channel values of the first row of the image.

The labels are encoded as integers in the range 0-9, corresponding to the numbered classes listed above
```

````python
import matplotlib.pylab as plt
import skimage
from sklearn.datasets import fetch_openml
import numpy as np
````

````python
cifar10 = fetch_openml(name='CIFAR_10_small')
````

````python
X = np.array(cifar10.data, dtype=np.int)
y = np.array(cifar10.target, dtype=np.int)
````

````python
X.shape, y.shape
````

````python
def show_example(X, index):
    img = X[index,:].reshape(3, 32, 32)
    img = np.transpose(img, [1, 2, 0])
    plt.imshow(img)
````

````python
plt.figure(figsize=(15,12))
for i in range(16):
    plt.subplot(4,4,i+1)
    show_example(X, i)
    plt.title(f"category: {y[i]}")
plt.show()
````

````python
plt.figure()
plt.hist(y)
plt.title("Histogram of class labels")
plt.xlabel("Category index")
plt.ylabel("Number of examples")
plt.show()
````
